#Animations

```python
import numpy as np

from einsteinpy.geodesic import Timelike
from astropy import units as u
from einsteinpy.plotting import StaticGeodesicPlotter

position = [4, np.pi/3, 0.]
momentum = [0., 1, 2.]
a = 0.99
steps = 100.
delta = 0.5

geod = Timelike(
    metric="Kerr",
    metric_params=(a,),
    position=position,
    momentum=momentum,
    steps=steps,
    delta=delta,
    return_cartesian=True
)

sgpl = StaticGeodesicPlotter()
sgpl.animate(geod, interval=2)
sgpl.show()
```
