```python
#1: importar as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt

#2: definir uma função que gere a gaussinada para valores específicos de a, b e x. Na física é mais comum representar gaussiana como funções de densidade de probabilidade
def gaussiana(x, mu, sigma):
    return ((1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2))
    #a fórmula acima é a de uma função de distribuição gaussiana

#3: definir valores para a média e o desvio padrão
mu = 0
sigma = 1

#intervalo dos eixos
x = np.linspace (-5, 5, 1000)
y = gaussiana(x, mu, sigma)

# Plotando a Gaussiana
plt.plot(x, y, label=f'$\mu={mu}$, $\sigma={sigma}$')
plt.xlabel('x') #legenda do eixo x
plt.ylabel('f(x)') #legenda do eixo y
plt.title('Distribuição Gaussiana') #título do gráfico
plt.legend() #legenda
plt.grid() #grade
plt.show() #exibir
```

Função sinc(x)

```python
import numpy as np
import matplotlib.pyplot as plt

def sinc(x):
    return(np.sin(x/np.pi))

x = np.linspace(-30, 30, 1000)
y = sinc(x)



# Plotando a função sinc
plt.plot(x, y, label=r'$sinc(x) = \frac{\sin(x)}{x}$')
plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.title('Função Sinc')
plt.legend()
plt.grid()
plt.show()
```

função exponencial

```python
import numpy as np
import matplotlib.pyplot as plt

def exponencial(x, a):
    return np.exp(a * x)

#definindo o eixo x
x = np.linspace(-5, 10, 1000)

# Valores de a
a_values = [1]

# Plotando as funções exponenciais
plt.figure(figsize=(8, 8))
for a in a_values:
    y = exponencial(x, a)
    plt.plot(x, y, label=f'$a = {a}$')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Exponencial')
plt.legend()
plt.grid()
plt.show()
```

Baskhara



```python
import numpy as np
import matplotlib.pyplot as plt

#calcular as raízes usando a fórmula de Bhaskara
def bhaskara(a, b, c):
    delta = b**2 - 4*a*c
    r1 = (-b + np.sqrt(delta)) / (2*a) * (delta > 0)  # Se delta > 0, r1 é calculado
    r2 = (-b - np.sqrt(delta)) / (2*a) * (delta > 0)  # Se delta > 0, r2 é calculado
    r = -b / (2*a) * (delta == 0)  # Se delta == 0, uma raiz unica r é calculada
    return r1, r2, r

# Função para plotar o gráfico
def plot_parabola(a, b, c):
    x = np.linspace(-20, 20, 400)
    y = a*x**2 + b*x + c
    plt.plot(x, y, label=f'Parábola: {a}x² + {b}x + {c}')

    # Calculando as raízes
    r1, r2, r = bhaskara(a, b, c)

    #A funçao plt.scatter serve para criar gráficos com dispersão ou pontos individuais
    plt.scatter(r1, 0, color='red', zorder=3)
    plt.scatter(r2, 0, color='red', zorder=3)
    plt.scatter(r, 0, color='red', zorder=3)

    #Definindo os eixos x e y e respectivas legendas
    plt.axhline(0, color='black',linewidth=3) #a função ahvline desenha a linha horizontal do gráfico
    plt.axvline(0, color='black',linewidth=3) #a função axvline desenha a linha vertical do gráfico
    #neste exercício escolhi não incluir a função "grid" para melhor desempenho do código
    plt.show()

# Definindo três casos, 1- delta positivo. 2- delta = 0, 3-delta negativo
casos = [
    (1, -3, 2),
    (1, 2, 1),
    (1, 1, 1)
]

# Plotando para cada caso
for a, b, c in casos:
    plot_parabola(a, b, c)
```

- Salvando arquivos

As bibliotecas mais adequadas para este caso seriam a numpy.save e numpy.savez, a diferença entre elas é que a np.savez compacta os dados em um único arquivo, desta forma os dados dos valores de x e y podem ser armazenados em um mesmo arquivo, euquanto a numpy.save necessita da criação de dois arquivos diferentes para os dados de x e y. Como estamos trabalhando com análise de dados na física, optei por usar o numpy.save, pois apesar de criar arquivos diferentes, ele mantém a precisão dos números. Para trabalhos com um grande volume de dados e muitas variáveis o mais adequado seria o numpy.savez.

```python
import numpy as np

#Definindo as constantes da equação
a, b, c = 3, 2, 4

#Criando um conjunto de valores para x
x = np.linspace(-5, 5, 100)

#definindo y (funçao dada na questão anterior)
y = a*x**2 + b*x + c

#Para salvar os dados em um arquivo, utilizando a função np.save:
np.save('dados_x.npy', x)
np.save('dados_y.npy', y)

```

```python
import numpy as np
import matplotlib.pyplot as plt

#para carregar os dados depois, utilizamos a função np.load
x = np.load("dados_x.npy")
y = np.load("dados_y.npy")

#para criar o gráfico utilizamos o procedimento padrão dos exercícios anteriores
plt.plot(x, y, color = 'pink')
plt.grid()
plt.show()



```

-Deltoid

```python
import numpy as np
import matplotlib.pyplot as plt

#parametro da curva (pode variar)
a = 1

#intervalor do parâmetro t
t = np.linspace(0, 2*np.pi, 100)

#Parametrização da curva deltoid
x = 2*a*np.cos(t) + a*np.cos(2*t)
y = 2*a*np.sin(t) - a*np.sin(2*t)

#Plotando a curva
plt.plot(x, y, color = 'pink')

#Definindo eixos
plt.axhline(0, color='black',linewidth=0.1)
plt.axvline(0, color='black',linewidth=0.1)
plt.xlabel ("Eixo x")
plt.ylabel("Eixo y")
plt.title("Deltoid")

plt.show()
```

**Espiral de Galileu***

A espiral de Galileu é definida pela paramartrização $r = \theta^{2}$
para este caso, o gráfico será plotado em coordenadas polares

```python
import numpy as np
import matplotlib.pyplot as plt

#definindo o intervalo de $\theta$
theta = np.linspace(0, 4*np.pi, 100)

#Parametrização da espiral de Galileu
r = theta**2

#Criando o gráfico em coordenadas polares
plt.subplot(projection ='polar')
plt.plot(theta, r, color = 'purple')
plt.title("Espiral de Galileu")
plt.show()
```

**Fey's function**

```python
import numpy as np
import matplotlib.pyplot as plt

#intervalo de theta
theta = np.linspace(0, 2*np.pi, 500)

#parametrização da função
r = np.exp(np.cos(theta)) - 2*np.cos(4*theta) + np.sin(theta/12)

#criando os gráficos, também em polares pos nessa função temos r em função de theta
plt.subplot(projection ='polar')
plt.plot(theta, r, color = 'pink')
plt.title("Fey's function")
plt.show()
```
