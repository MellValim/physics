```python
#1: importar as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt

#2: definir uma função que gere a gaussinada para valores específicos de a, b e x. Na física é mais comum representar gaussiana como funções de densidade de probabilidade
def gaussiana(x, mu, sigma):
    return ((1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2))
    #a fórmula acima é a de uma função de distribuição gaussiana

#3: definir valores para a média e o desvio padrão
mu = 0
sigma = 1

#intervalo dos eixos
x = np.linspace (-5, 5, 1000)
y = gaussiana(x, mu, sigma)

# Plotando a Gaussiana
plt.plot(x, y, label=f'$\mu={mu}$, $\sigma={sigma}$')
plt.xlabel('x') #legenda do eixo x
plt.ylabel('f(x)') #legenda do eixo y
plt.title('Distribuição Gaussiana') #título do gráfico
plt.legend() #legenda
plt.grid() #grade
plt.show() #exibir
```

Função sinc(x)

```python
import numpy as np
import matplotlib.pyplot as plt

def sinc(x):
    return(np.sin(x/np.pi))

x = np.linspace(-30, 30, 1000)
y = sinc(x)



# Plotando a função sinc
plt.plot(x, y, label=r'$sinc(x) = \frac{\sin(x)}{x}$')
plt.xlabel('x')
plt.ylabel('sinc(x)')
plt.title('Função Sinc')
plt.legend()
plt.grid()
plt.show()
```

função exponencial

```python
import numpy as np
import matplotlib.pyplot as plt

def exponencial(x, a):
    return np.exp(a * x)

#definindo o eixo x
x = np.linspace(-5, 10, 1000)

# Valores de a
a_values = [1]

# Plotando as funções exponenciais
plt.figure(figsize=(8, 8))
for a in a_values:
    y = exponencial(x, a)
    plt.plot(x, y, label=f'$a = {a}$')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Exponencial')
plt.legend()
plt.grid()
plt.show()
```

Baskhara



```python
import numpy as np
import matplotlib.pyplot as plt

#calcular as raízes usando a fórmula de Bhaskara
def bhaskara(a, b, c):
    delta = b**2 - 4*a*c
    r1 = (-b + np.sqrt(delta)) / (2*a) * (delta > 0)  # Se delta > 0, r1 é calculado
    r2 = (-b - np.sqrt(delta)) / (2*a) * (delta > 0)  # Se delta > 0, r2 é calculado
    r = -b / (2*a) * (delta == 0)  # Se delta == 0, uma raiz unica r é calculada
    return r1, r2, r

# Função para plotar o gráfico
def plot_parabola(a, b, c):
    x = np.linspace(-20, 20, 400)
    y = a*x**2 + b*x + c
    plt.plot(x, y, label=f'Parábola: {a}x² + {b}x + {c}')

    # Calculando as raízes
    r1, r2, r = bhaskara(a, b, c)

    #A funçao plt.scatter serve para criar gráficos com dispersão ou pontos individuais
    plt.scatter(r1, 0, color='red', zorder=3)
    plt.scatter(r2, 0, color='red', zorder=3)
    plt.scatter(r, 0, color='red', zorder=3)

    #Definindo os eixos x e y e respectivas legendas
    plt.axhline(0, color='black',linewidth=3) #a função ahvline desenha a linha horizontal do gráfico
    plt.axvline(0, color='black',linewidth=3) #a função axvline desenha a linha vertical do gráfico
    #neste exercício escolhi não incluir a função "grid" para melhor desempenho do código
    plt.show()

# Definindo três casos, 1- delta positivo. 2- delta = 0, 3-delta negativo
casos = [
    (1, -3, 2),
    (1, 2, 1),
    (1, 1, 1)
]

# Plotando para cada caso
for a, b, c in casos:
    plot_parabola(a, b, c)
```
